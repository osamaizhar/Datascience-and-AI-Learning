# Mean , Max, Min, Standard deviation (Std) , Variance 

**Variance** = measure of the average degree to which each number is different to the mean.
It represents how spread out the data is from the mean.

**High Variance** = wider range of numbers
**Low Variance** = lower range of numbers


**Standard Deviation** = the square root of the **variance**. It's a measure of how spread out a group of numbers is from the mean. A low standard deviation indicates that the values tend to be close to the mean, while a high standard deviation indicates that the values are spread out over a wider range.

In other words, **variance** is a measure of the *"average distance"* between each data point and the mean, while **standard deviation** is a measure of the *"typical distance"* between each data point and the mean.

```python
print(f" A2: {a2}")
print(f"Mean : {np.mean(a2)}")
print(f"Min : {np.min(a2)}")
print(f"Max : {np.max(a2)}")
print(f"STD : {np.std(a2)}")
print(f"VAR : {np.var(a2)}")
```

```python
data = [1, 2, 3, 4, 5]
mean = np.mean(data)
variance = np.var(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

# Standard Deviation  = square root of Variance 

```python
np.sqrt(np.var(data))
```

# Reshape array using .reshape(rows,columns)


**Reshaping Arrays**

Reshaping an array means changing its shape or dimensions without changing the data itself. In NumPy, you can use the `reshape()` function to reshape an array.

**Here's the general syntax:**

`array.reshape(new_shape)`
Where new_shape is a tuple representing the new shape of the array.

**How Reshaping Works**

When you reshape an array, NumPy will attempt to rearrange the elements to fit the new shape. Here are the rules:

Total number of elements remains the same: The total number of elements in the original array must be equal to the total number of elements in the new shape.
Elements are rearranged in row-major order: NumPy rearranges the elements in row-major order (i.e., from left to right, top to bottom).
New shape must be compatible: The new shape must be compatible with the original array's shape. For example, if the original array has 12 elements, the new shape must also have 12 elements.

```python
print(a2.shape)
print(a2.reshape(2, 3, 1).shape)
# when there are 3 dim , first one is no of layers , 2nd one is rows,  3rd one is columns
a2_reshaped = a2.reshape(2, 3, 1)
print(a2_reshaped)
```

# Nut butter Sales Problem

Calculating the total sales using data analysis and dot product

```python
# ^ Creating the sales array
sales = np.array([[2, 7, 1], [9, 4, 16], [11, 14, 18],
                 [13, 13, 16], [15, 18, 9]])
sales.shape
```

```python
# ^ Creating a dataframe
import pandas as pd
weekly_sales = pd.DataFrame(sales, index=["Mon", "Tue", "Wed", "Thurs", "Fri"], columns=[
                            "Almond Butter", "Peanut Butter", "Cashew Butter"])
weekly_sales
```

```python
# ^ Creating the prices array
prices = np.array([10, 8, 12])
prices
```

```python
# ^ Creating butter prices dataframe

'''
Returns error ValueError: Shape of passed values is (3, 1), indices imply (1, 3) since prices 
is (3,) 1d array with 3 items. But the dataframe we have has 1 index and 3 columns so (3,1) 2d array with 1 row and 3 columns is expected for prices
'''
butter_prices = pd.DataFrame(prices, index=["Price"], columns=[
                             "Almond Butter", "Peanut Butter", "Cashew Butter"])
butter_prices
```

```python
# ^ Therefore prices needs to be reshaped to (1,3)
print(prices.shape)  # returns (3,) which is 1d array of 3 items
print("1d array of 3 items", prices)
print("2d array of 1 row and 3 columns", prices.reshape(1, 3))
butter_prices = pd.DataFrame(prices.reshape(1, 3), index=["Price"], columns=[
                             "Almond Butter", "Peanut Butter", "Cashew Butter"])
butter_prices
```

```python
# ^ Finding total sales
total_sales = prices.dot(sales)
```

```python
# ^ Resolving shape error using transpose
print(prices.shape)  # (3,)
# (5,3)   as we can see we should transpose sales so that we can get dot product to get total sales
print(sales.shape)
print(sales)
print("sales.T", sales.T)
print("sales Reshape", sales.reshape(3, 5))
# reshape will not work here since it reshape row wise and transpose  replaces row with column
total_sales = prices.dot(sales.T)
print(total_sales)  # exact same as the excel sheet result
```

```python
# ^ Adding the total sales as a column into the weekly_sales dataframe

weekly_sales["$ Total"] = total_sales
weekly_sales
```
