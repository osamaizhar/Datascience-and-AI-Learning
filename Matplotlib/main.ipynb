```python
# Method 2
'''
The four numbers in the list are:

[left, bottom, width, height]
These numbers define the position and size of the new axes in the figure.

left: The x-coordinate of the left edge of the axes.
bottom: The y-coordinate of the bottom edge of the axes.
width: The width of the axes.
height: The height of the axes.
'''
fig = plt.figure()  # creates a figure
ax = fig.add_axes([1, 1, 1, 1])  # creates axes
ax.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

# prepare data
x = [0, 1, 2, 3]
y = [11, 22, 33, 44]

# Setup plot
fig, ax = plt.subplots(figsize=(10, 10))  # (width, height)

# Plot data , since we are not specifying the variation between axis values it is 0.5 by default
ax.plot(x, y)

# Setting title ana labels
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# Save & show (you save the whole figure)
fig.savefig("sample-plot.png")
plt.show()
```

# Bar plot using Dictionary


```python
nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())  # (x, y)
ax.set(title="Dan's Nut Butter Store",
       ylabel="Price ($)")
```

# Horizontal Bar Graph


```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values())
```

# Histogram Plot (Bell Curve via Normal Distribution)


```python
# generates 1000 random values , randn: Generates random numbers from a standard normal distribution (mean 0, standard deviation 1).
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

# Two options for Subplot

Option 1: Making all axes with separate vars


```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,  # ^ Controls rows and columns for the figure
                                             figsize=(10, 5)
                                             )
# * plot each to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

Option 2: Using indexes to access each sub axis


```python
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```

# Which one should you use ? (Pyplot vs Matplotlib OO method)

- when plotting something quickly better to use **pyplot**
- when plotting something more advanced use **OO method**


```python
heart_disease
```

```python
# getting row of data where age is greater than 50
over_50 = heart_disease[heart_disease.age > 50]
over_50
```

# Pyplot Method


```python
# plotting on the dataframe , c is for color
over_50.plot(kind="scatter", x="age", y="chol", c="target")
```

# OO Method


```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)

ax.set_xlim([45, 100])  # sets limit on x axis from 45 to 100
```

```python
over_50.target.values
```

# Creating Colored Scatter Plot for Heart Disease

OO method from scratch


```python

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol Levels")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")
```

# Adding horizontal line to graph to show average cholesterol level


```python
###  ----------------Same Code as before ---- ------------------###
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol Levels")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# adding the horizontal line for average cholesterol
ax.axhline(over_50["chol"].mean(), linestyle="--", color="red")
```

# Multiple Plots on the same figure


```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               sharex=True,  # both plots share the same axis
                               figsize=(10, 8))

# ----------------------- Axis 0 : Heart Disease and Cholesterol level ----------------------------------

# Add data for ax0
scatter = ax0.scatter(over_50['age'],
                      over_50["chol"],
                      c=over_50["target"])

# Customize ax0 , Set title ,labels and legend
ax0.set(title="Heart Disease and Cholesterol Levels",
        xlabel="Age",
        ylabel="Cholesterol Levels")

ax0.legend(*scatter.legend_elements(), title="Target")

# Set a mean line
ax0.axhline(y=over_50['chol'].mean(),
            color='b',
            linestyle='--',
            label='Average')

# ---------- Axis 1: Heart Disease and Max Heart Rate Levels ----------

# Add data for  ax1
scatter = ax1.scatter(over_50['age'],
over_50['thalach'],
c=over_50['target'])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
xlabel="Age",
ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(),title="Target")

# Setup a mean line 

ax1.axhline(y=over_50['thalach'].mean(),
color ="b",
linestyle= "--",
label = "Average")

# Title the figure

fig.suptitle("Heart Disease Analysis",
fontsize=16,
fontweight= "bold")

```
